### Проверка формулы вычисления эффективности обогащения

```python
def calc_recovery(C, F, T):
    return (100*(C*(F-T))/(F*(C-T))) 
```

```python
# для расчета используем данные без пропусков (иначе возникает ошибка при расчете MAE)
df_rec_calc = df_train.loc[~df_train['rougher.output.recovery'].isna(),
                       ['rougher.output.concentrate_au','rougher.input.feed_au','rougher.output.tail_au','rougher.output.recovery']]

rec_temp = calc_recovery(df_rec_calc['rougher.output.concentrate_au'],
                        df_rec_calc['rougher.input.feed_au'],
                        df_rec_calc['rougher.output.tail_au'],)

rec_mae = mean_absolute_error(rec_temp, df_rec_calc['rougher.output.recovery'])
print(rec_mae)
```

Значение МАЕ крайне мало, а значит формула корректна. Вдальнейшем можно попробовать ее применить для заполнения пропусков.

### Исследование изменения концентрации элементов на каждом этапе

Рассчитаем среднее значение концентраций золота, серебра и свинца на каждой из стадий обработки.

```python
print('<<<<<<<<<РАСЧЕТ КОНЦЕНТРАЦИЙ МЕТАЛЛОВ>>>>>>>>')

print("1. Исходное сырье")
au = df_full['rougher.input.feed_au'].mean().round(2)
ag = df_full['rougher.input.feed_ag'].mean().round(2)
pb = df_full['rougher.input.feed_pb'].mean().round(2)
print('   концентрация золота =', au)
print('   концентрация серебра =', ag)
print('   концентрация свинца =', pb)
row = {'stadia':['1.Init']*3, 'type':['au','ag','pb'], 'conc':[au,ag,pb]}
df_temp = pd.DataFrame(data=row)

print("2. После флотации")
au = df_full['rougher.output.concentrate_au'].mean().round(2)
ag = df_full['rougher.output.concentrate_ag'].mean().round(2)
pb = df_full['rougher.output.concentrate_pb'].mean().round(2)
print('   концентрация золота =', au)
print('   концентрация серебра =', ag)
print('   концентрация свинца =', pb)
row = {'stadia':['2.Rougher']*3, 'type':['au','ag','pb'], 'conc':[au,ag,pb]}
df_temp = df_temp.append(pd.DataFrame(row), ignore_index=True)

print("3. После первого этапа очистки")
au = df_full['primary_cleaner.output.concentrate_au'].mean().round(2)
ag = df_full['primary_cleaner.output.concentrate_ag'].mean().round(2)
pb = df_full['primary_cleaner.output.concentrate_pb'].mean().round(2)
print('   концентрация золота =', au)
print('   концентрация серебра =', ag)
print('   концентрация свинца =', pb)
row = {'stadia':['3.Cleaner-1']*3, 'type':['au','ag','pb'], 'conc':[au,ag,pb]}
df_temp = df_temp.append(pd.DataFrame(row), ignore_index=True)

print("4. После второго этапа очистки (в финальном концентрате)")
au = df_full['final.output.concentrate_au'].mean().round(2)
ag = df_full['final.output.concentrate_ag'].mean().round(2)
pb = df_full['final.output.concentrate_pb'].mean().round(2)
print('   концентрация золота =', au)
print('   концентрация серебра =', ag)
print('   концентрация свинца =', pb)
row = {'stadia':['4.Cleaner-2']*3, 'type':['au','ag','pb'], 'conc':[au,ag,pb]}
df_temp = df_temp.append(pd.DataFrame(row), ignore_index=True)
```

```python
sb.barplot(x='stadia',y='conc',hue='type',data=df_temp)
```

Концентрация золота растет от стадии к стадии - это самое главное.  
Концентрация серебра после роста на стадии флотации затем уменьшается.  
Концентрация свинца незначительно увеличивается от стадии к стадии.

```python
def hist_by_stadia(df, col_list, figsize=[16,8], bins=30):
    ax = df.plot(kind='hist', y=col_list[0], alpha = 0.5, bins = bins, figsize=figsize, linewidth = 2, density=True)
    for i in range(1,len(col_list)):
        df.plot(kind='hist', y=col_list[i], ax=ax, alpha = 0.5, bins = bins, linewidth = 2, density=True)     
    ax.legend(col_list)
```

## Модель

### Функция для вычисления итогового *sMAPE*

```python
def sMAPE(y_target, y_pred):
    return sum(abs(y_target - y_pred)/((abs(y_target) + abs(y_pred))/2))*(1/len(y_target))*100
 
def total_sMAPE(smape_r, smape_f):
    return 0.25*smape_r + 0.75*smape_f
 
#создадим свою метрику для выбора лучшей модели на этапе кросс-валидации
sMAPE_score = make_scorer(sMAPE, greater_is_better=False)
```

### Обучение и проверка моделей

Подготовим выборки для обучения моделей

```python
rnd_state=42

features_train = df_train.drop(['rougher.output.recovery','final.output.recovery'], axis=1)
features_test = df_test.drop(['rougher.output.recovery','final.output.recovery'], axis=1)

scaler = StandardScaler()
scaler.fit(features_train)
features_train = scaler.transform(features_train)
features_test = scaler.transform(features_test)

target_train_r = df_train['rougher.output.recovery']
target_train_f = df_train['final.output.recovery']

target_test_r = df_test['rougher.output.recovery']
target_test_f = df_test['final.output.recovery']
```

Для удобства вместо cross_val_score будем использовать GridSearchCV, который позволяет помимо кросс-валидации сразу перебрать гиперпараметры модели

#### Линейная регрессия

```python
def model_result_show(model_r, model_f, model_name):
    sMape_r = -1*model_r.best_score_
    sMape_f = -1*model_f.best_score_

    print(model_name)
    print('-----------------------------------------')
    print('Гиперпараметры лучшей модели rougher: ', model_r.best_params_)
    print('Гиперпараметры лучшей модели final: ', model_f.best_params_)
    print('-----------------------------------------')
    print('sMAPE rougher на обучающей выборке:', sMape_r)
    print('sMAPE final на обучающей выборке: ', sMape_f)
    print('sMAPE total на обучающей выборке: ', total_sMAPE(sMape_r, sMape_f))
    print('-----------------------------------------')
    
```

```python
%%time

model = LinearRegression()
param_grid = {}

model_r=GridSearchCV(model, param_grid, scoring=sMAPE_score, cv=5)
model_r.fit(features_train, target_train_r)

model_f=GridSearchCV(model, param_grid, scoring=sMAPE_score, cv=5)
model_f.fit(features_train, target_train_f)

model_result_show(model_r, model_f,'Линейная регрессия')
```

#### Дерево решений

```python
%%time

model = DecisionTreeRegressor(random_state=rnd_state)
param_grid = {'max_depth': range(1,11,1)}

model_r=GridSearchCV(model, param_grid, scoring=sMAPE_score, cv=5)
model_r.fit(features_train, target_train_r)

model_f=GridSearchCV(model, param_grid, scoring=sMAPE_score, cv=5)
model_f.fit(features_train, target_train_f)

model_result_show(model_r, model_f,'Дерево решений')
```

#### Случайный лес

```python
%%time

model = RandomForestRegressor(random_state=rnd_state)
param_grid = {
    'n_estimators': [50],
    'max_depth': [10], #считает слишком долго - оставил уже рассчитанный вариант range(1, 11, 1),
}

model_r=GridSearchCV(model, param_grid, scoring=sMAPE_score, cv=5)
model_r.fit(features_train, target_train_r)

model_f=GridSearchCV(model, param_grid, scoring=sMAPE_score, cv=5)
model_f.fit(features_train, target_train_f)

model_result_show(model_r, model_f,'Случайный лес')
```

#### Catboost

```python
%%time

model = CatBoostRegressor(random_state=rnd_state)
#модель настолько хороша и параметры по умолчанию подобраны настолько хорошо что их можно не менять
param_grid = {}

model_r=GridSearchCV(model, param_grid, scoring=sMAPE_score, cv=5)
model_r.fit(features_train, target_train_r)

model_f=GridSearchCV(model, param_grid, scoring=sMAPE_score, cv=5)
model_f.fit(features_train, target_train_f)

model_result_show(model_r, model_f,'CatBoostRegressor')
```

```python
#сохраняем лучшие модели для последующего тестирования на тестовой выборке
best_model_r = model_r
best_model_f = model_f
```

### Тестирование лучшей модели на тестовой выборке

```python
model_r_pred = best_model_r.predict(features_test)
model_f_pred = best_model_f.predict(features_test)

sMape_r = sMAPE(target_test_r, model_r_pred)
sMape_f = sMAPE(target_test_f, model_f_pred)

print('Лучшая модель: CatBoostRegressor')
print('-----------------------------------------')
print('Гиперпараметры лучшей модели rougher: ', best_model_r.best_params_)
print('Гиперпараметры лучшей модели final: ', best_model_f.best_params_)
print('-----------------------------------------')
print('sMAPE rougher на тестовой выборке:', sMape_r)
print('sMAPE final на тестовой выборке: ', sMape_f)
print('sMAPE total на тестовой выборке: ', total_sMAPE(sMape_r, sMape_f))
print('-----------------------------------------')
```
