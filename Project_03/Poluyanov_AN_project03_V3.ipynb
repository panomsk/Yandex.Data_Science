# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных**
- airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
- balcony — число балконов
- ceiling_height — высота потолков (м)
- cityCenters_nearest — расстояние до центра города (м)
- days_exposition — сколько дней было размещено объявление (от публикации до снятия)
- first_day_exposition — дата публикации
- floor — этаж
- floors_total — всего этажей в доме
- is_apartment — апартаменты (булев тип)
- kitchen_area — площадь кухни в квадратных метрах (м²)
- last_price — цена на момент снятия с публикации
- living_area — жилая площадь в квадратных метрах (м²)
- locality_name — название населённого пункта
- open_plan — свободная планировка (булев тип)
- parks_around3000 — число парков в радиусе 3 км
- parks_nearest — расстояние до ближайшего парка (м)
- ponds_around3000 — число водоёмов в радиусе 3 км
- ponds_nearest — расстояние до ближайшего водоёма (м)
- rooms — число комнат
- studio — квартира-студия (булев тип)
- total_area — площадь квартиры в квадратных метрах (м²)
- total_images — число фотографий квартиры в объявлении

На этапе предобработки данных заполним пропущенные значения и изменим (при необходимости) тип данных столбцов.

Определим функцию для вывода информации о пропусках по наименованию столбца

```python
def isna_info_col(data, col):
    print('Пропущено значений в столбце ',col,'-', data[col].isna().sum())
    print('Процент пропусков от общего количества строк: {:.2%}'.format(data[col].isna().sum()/data[col].shape[0]))


isna_info_col(df,'ceiling_height')    
df['ceiling_height'].describe()
```

В столбце пропущено 38% значений, поэтому удалить строки с пропусками нельзя. Выделить группы квартир, по которым можно было бы заменить значения на медианное, на первый взгляд не получается. Кроме того в столбце явно присутствуют выбросы (1, 100). 

```python
plt.subplots(figsize=(20,15))
sb.heatmap(df.corr(), cmap="YlGnBu", annot=True)
```

Корреляция параметра ceiling_height другими числовыми параметрами отсутствует. Но зависимость может быть нелинейной, поэтому рассмотрим таблицу медианных значений высоты потолков в зависимости от этажности дома, ранее были распространены типовые 5, 9 этажные дома с низкой высотой потолков, в современном строительстве потолки обычно имеют высоту 2,7 - 2,8 метров. Проверим предположения.

```python
locality_ceiling_height = df.groupby('floors_total')['ceiling_height'].agg(['median', 'count'])
locality_ceiling_height.sort_values(by='floors_total', inplace=True)
locality_ceiling_height.head(10)
```

Зависимость есть! Заполним пропуски медианным значением в зависимости от этажности дома

```python
data_transformed = df.groupby('floors_total')['ceiling_height'].transform('median')
df['ceiling_height'] = df['ceiling_height'].fillna(data_transformed)
isna_info_col(df,'ceiling_height')
```

Пропусков осталось менее 1%, удалим данные строки

```python
df.dropna(subset=['ceiling_height'], inplace=True)
```

floors_total            23613 non-null float64

Проанализируем пропуски, заменим тип данных на int64

```python
isna_info_col(df,'floors_total')
df['floors_total'].value_counts()
```

Корректно заполнить пропущенные значения по имеющимся данным возможности нет, помог бы адрес дома для заполнения этажности домов с одинаковым адресом, но у нас такой информации нет. Этажность используется для дальнейшего анализа (влияет на определение последний или непоследний этаж), поэтому необоснованное заполнение пропусков может повлиять на выводы. Пропущенных значений меньше 1%, поэтому можно удалить пропуски.

```python
df.dropna(subset=['floors_total'], inplace=True)

try:
    df['floors_total'] = df['floors_total'].astype('int64')
except:
    print('Ошибка преобразования типов')
    
isna_info_col(df,'floors_total')
```

living_area             21796 non-null float64

Тип данных корректный. проанализируем пропуски

```python
isna_info_col(df,'living_area')
```

Пропусков порядка 8%, поэтому удалить строки не получится. 
Выше построили матрицу корреляции по датафрецму и видна сильная положительная корреляция (0,94) между жилой и общей площадью. Поэтому заполним жилую площадь, как средний коэффициент * общую площадь

```python
data_transformed = (df['living_area'] / df['total_area']).median() * df['total_area']
df['living_area'] = df['living_area'].fillna(data_transformed)
isna_info_col(df,'living_area')
```

Тип данных приведем к булевому (т.к. возможны только значения True или False). Проанализируем пропуски.

```python
isna_info_col(df,'is_apartment')
df['is_apartment'].value_counts()
```

Как мы видим пропусков почти 90%. Среди заполненных значений преобладают значения False. Можно предположить, что для незаполненных значений корректным является значение False, поскольку доля апартаментов среди недвижимости небольшая.
Заполним пропуски значением False и преобразуем тип данных к bool

```python
df['is_apartment'].fillna(False, inplace=True)

try:
    df['is_apartment'] = df['is_apartment'].astype('bool')
except:
    print('Ошибка преобразования типов')
    
df['is_apartment'].value_counts()
```

Определим вспомогательную функцию, которая выведет необходимые данные по каждому из анализируемых столбцов (describe, boxplot, hist)

```python
def column_description(df, col, bins_count=30, range_min=0, range_max=0):
    display(df[col].describe())
    #рассчитаем квантили
    display(df[col].quantile([0.01,0.05,0.95,0.99]))
    
    #рассчитаем значения 'усов' в диаграмме с усами (будут использованы в пукте 4.2)
    Q1 = df[col].quantile(q=0.25)
    Q3 = df[col].quantile(q=0.75)
    IQR = Q3 - Q1
    print('Q1-1.5*IQR = ',Q1-1.5*IQR)
    print('Q3+1.5*IQR = ',Q3+1.5*IQR)
    
    if range_max==0: df[col].hist(bins=bins_count)
    else: df[col].hist(bins=bins_count, range=(range_min,range_max))
    plt.title(col)
    plt.show()
    df.boxplot(column=col)
    if range_max!=0: plt.ylim(range_min, range_max)
    plt.title(col)
    plt.show()
```

Определим вспомогательную функцию

```python
def show_corr_and_plot(df, col, corr_counting=True, need_line=True):
    if (corr_counting==True):
        print('Коэффициент корреляции price_sq_meter -',col,'=', round(df[col].corr(df['price_sq_meter']),2))
    pv_table = df.pivot_table(index=col, values='price_sq_meter', aggfunc=['median','mean'])
    
    style_str='-o'
    if need_line==False: style_str='o'
    
    pv_table.plot(y = ['mean','median'], style=style_str, grid=True, legend=True, label=['mean','median'], alpha=0.5)
    plt.show()
```
